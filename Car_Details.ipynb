```python
from selenium import webdriver
Page_Number = 2
CarName = []
Price = []

for CurrentPage in range(1,Page_Number+1):
    driver = webdriver.Chrome(executable_path=r"/home/haripriya/Downloads/chromedriver_linux64/chromedriver")

    driver.get('https://www.gotoauto.ca/inventory/?pg='+str(CurrentPage))
    html_Page_Number ='page'+str(CurrentPage)+'.html'
    with open(html_Page_Number, 'w+') as f:
        f.write(driver.page_source)
        f.close()
    driver.close()
    with open(html_Page_Number) as fp:
        soup = BeautifulSoup(fp)
        
    Car_List = soup.find_all(class_="list")
    for i in Car_List:
        print(i.find("h4",attrs={'class':'name desc_l5'}).text)
        CarName_Value = i.find("h4",attrs={'class':'name desc_l5'}).text
        CarName_Value_Update = CarName_Value.replace('\t'," ").replace('\n'," ")
        NewCarName_Value = " ".join(CarName_Value_Update.split())
        CarName.append(NewCarName_Value)

        print(i.find("span",attrs={'class':'price'}).text)
        Price.append(i.find("span",attrs={'class':'price'}).text)    
        
```
